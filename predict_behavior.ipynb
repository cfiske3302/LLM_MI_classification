```python
def get_message(conversation, prompt_path):
    conv = ""
    for utterance in conversation:
        conv += f"{utterance['interlocutor']}: {utterance['utterance_text']}\n"

    with open(prompt_path, 'r') as f:
        prompt = f.read()
    # print(therapist_utterances)
    message = prompt.format(conversation=str(conv))
    # print(message)
    formated_message = [{"role": "user", "content": message}]
```

```python
def predict_conversation(conv_path, output_path, prepend_num=4):
    potential_responses = ['ADP','ADW','AF','CO','DI','EC','FA','FI','GI','QUC','QUO','RCP','RCW','RES','REC','RF','SU','ST','WA']
    dict_list = read_dict_list(conv_path)
    file = codecs.open(output_path, 'w', 'utf-8')
    conversation = []
    for i, utterance in enumerate(dict_list):
        if i < prepend_num:
            utterance["llm_predict"] = "**prepend**"
            print(json.dumps(utterance), file=file)
            conversation.append(utterance)
            continue

        input_message = get_message(conversation=conversation, prompt_path='prompts/prompt_predict')
        reply = get_completion_from_messages(input_message)

        for pr in potential_responses:
            if pr in reply:
                reply = pr
                break
        utterance["llm_predict"] = reply
        print(json.dumps(utterance), file=file)
        conversation.append(utterance)
    file.close()
```
